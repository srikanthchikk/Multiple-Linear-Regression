# Multiple Linear Regression Computer_Data Problem Statement
Perform multilinear regression with price as the output variable and document the different RMSE values

Objective = to check the different RMSE values

```python
# Importing the required library  which are helps to bilding the model  
import numpy as np # for mathematical and logical operations and consisting of multidimensional array objects 
import pandas as pd # for Data Manipulation
#Graphical Representation
import matplotlib.pyplot as plt # mostly used for visualization purposes 
```

```python
#lets load the data set
computer = pd.read_csv("Computer_Data.csv")
computer.describe()
computer.columns
```

```python
#as the first column is given serial number, so this is not required
#we will drop this column Unnamed: 0
computer = computer.drop('Unnamed: 0', axis=1)
computer
```

```python
#lets make the dependant and independant variables
#as from the data set profit is depends on the remaining all
X = computer.iloc[:,1:10] #independ variables excluding price 
y = computer.iloc[:,0]  #depended variable price
```

```python
#as the State column is catergorical we will convert this using onehot encoding using get_dummies()
#the impact of first NewYork is removed
cd = pd.get_dummies(X['cd'])
multi = pd.get_dummies(X['multi'])
premium = pd.get_dummies(X['premium'])
```

```python
#lets remove the categorical column
X = X.drop(['cd','multi','premium'], axis=1)
X
```

```python
#lets use concatination
X = pd.concat([X,cd,multi,premium], axis=1)
X
```

```python
#let's split the data into train and test
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
#now lets build a model using fitting Multilinear Regression to the training set
from sklearn.linear_model import LinearRegression
start_f = LinearRegression() 
start_f.fit(X_train, y_train)
```

```python
#prediction for the test results
y_pred = start_f.predict(X_test)
y_pred
```

```python
#for comarision to check whether the y_test is good or y_pred is good
#we will use r^2 score from the sklearn package
from sklearn.metrics import r2_score
score = r2_score(y_test, y_pred) #78.4277% ## as this is near to 1 so the model is good
score
```

```python
# Finding the MAE, MSE, RMSE socre 
from sklearn import metrics
print('MAE:', metrics.mean_absolute_error(y_test, y_pred))
print('MSE:', metrics.mean_squared_error(y_test, y_pred))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```
